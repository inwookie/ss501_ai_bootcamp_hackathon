```python
# 다듬어진 속성인 Trim_table을 불러오고 확인한다
data = pd.read_csv(r'Trim_table.csv')
car = data.copy()
car.head()
```

* Maker : 차의 브랜드, 메이커

* Trim : 옵션을 포함한 차의 모델명

* Price : 그 연도에 판매된 가격

* Fuel_type : 연료의 타입
* Engine_size : 엔진 사이즈

```python
car.info()
```

지금 상태로는 null 값은 없다.
여기에 Entry_price를 추가하고 싶다

```python
car['Entry_price'] = np.nan
```

```python
# 이제 Entry_price라는 컬럼이 생겼다!
car.info()
```

이제 올바른 브랜드와 종류, 연도에 따른 Entry_price값을 적절하게 넣어보자 <br>
Trim_table의 처음부터 끝까지 읽으면서 <br>
Genmodel_ID, Year이 같은 차의 Entry_price를 찾아서 넣어준다

```python
count = 0
for i in range(0, 335562):
    c = car.loc[i]
    condition = (price['Genmodel_ID'] == c['Genmodel_ID']) & (price['Year'] == c['Year'])
    filtered = price.loc[condition]
    car.iat[i, 9] = filtered['Entry_price']
    count += 1
    if count % 1000 == 0:
        print(f'{count}개 완료')
print('모두 완료!')
```

```python
car = car.astype({'Entry_price':'int'})
car.info()
```

null 없이 잘 채워졌다
